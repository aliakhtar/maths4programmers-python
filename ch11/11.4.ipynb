Exercise 11.4: Plot the cross section of

where y = 1. Then plot the cross section of h(x, y), where x = pi / 6

```python
from math import e, pi, sin
h = lambda x,y: (e ** y) * sin(x)

args = list(range(-5, 5))
values = [h(x, 1) for x in args]
plt.plot(args, values)
```

Exercise 11.5: What are the partial derivatives of the function h(x, y) from the first exercise? What is the gradient? What
is the value of the gradient at (x, y) = (ππ/6, 1)?

```python
from sympy import *
from sympy.core.core import *
x = Symbol('x')
y = Symbol('y')
x_g = ((e ** y) * sin(x)).diff(x)
x_g
```

```python
y_g = ((e ** y) * sin(x)).diff(y)
y_g
```

```python
v = (x_g.subs(x, pi / 6).subs(y,1), y_g.subs(x, pi / 6).subs(y,1))
v
```
