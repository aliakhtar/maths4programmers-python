Mini-project 3.2: There are exactly eight 3D vectors whose coordinates are all either +1 or -1. For instance, (1, -1, 1) is
one of these. Plot all of these eight vectors as points. Then figure out how to connect them with line segments using
Segment3D objects to form the outline of a cube.
Hint: you’ll need 12 segments in total.

```python
values = [1, -1]
points = [(x, y, z) for x in values for y in values for z in values]
print(len(points))
for p in points:
    print(p)
```

```python
edges = [((-1, y, z), (1, y, z)) for y in values for z in values] + \
        [ ( (x, -1, z), (x, 1, z)) for x in values for z in values ] + \
        [ ( (x, y, -1), (x, y, 1) ) for x in values for y in values ]

print(len(edges))

segments = [ Segment3D(e[0], e[1]) for e in edges ]
draw3d(*segments, Points3D(*points))
```
